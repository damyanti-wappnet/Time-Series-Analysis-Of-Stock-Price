```python
# Required libraries for Stock Price Analysis

# Import datetime library from datetime module to work with date and time
from datetime import datetime

#import express library from plotly module for data visualization
import plotly.express as px
```

```python
#import the yahoo finance package to fetch real time data
import yfinance as yf

#download the stock data for Microsoft
#yf.download() will download stock data
""" 
parameters: -> first parameter will be the name of the company whose data we want to analyze (MSFT to get microsoft stock data)
            -> start will take starting date of data The format of the date is YYYY-MM-DD.
            -> end will take ending date of data The format of the date is YYYY-MM-DD.
"""
MSFT_stock_data = yf.download('MSFT', start="2020-01-01", end="2022-11-30") 
```

```python
#  Line graph visualizing the trends in the close prices of Microsoft (with filters for time period of 1m, 6m, YTD, 1y, all)
"""
px.line() will plot line graph

parameters  -> First parameter will be dataset
            -> x will take values for x-axis
            -> y will take values for y-axis
            -> title will hold the title of graph
"""
figure = px.line(MSFT_stock_data, x = MSFT_stock_data.index, y = 'Close', title = 'Close price Analysis with Time Period ')

# update_xaxes() is used for updating the x-axis properties
# rangeslider_visible = True -> will put slider to arrange time duration
figure.update_xaxes(rangeslider_visible=True)

# update_xaxes(rangeselector=dict()) is used for create a range selector on the x-axis
# The step parameter can be set to ( "month" | "year" | "day" | "hour" | "minute" | "second" | "all" ) 
# The stepmode parameter can be set to "backward" or "todate" to specify the direction of the range selector.

figure.update_xaxes(
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1m", step="month", stepmode="backward"), 
            dict(count=3, label="3m", step="month", stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=1, label="YTD", step="year", stepmode="todate"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(step="all")
        ])
    )
)


figure.show() #show graph
```

```python
# Bar chart for open, close, high, low trends in Microsoft prices(with filters for time period of 1m, 6m, YTD, 1y, all)

"""
px.line() will plot line graph

parameters : First parameter will be dataset
            -> x will take values for x-axis
            -> y will take values for y-axis
            -> title will hold the title of graph
"""

figure = px.bar(MSFT_stock_data, x = MSFT_stock_data.index, y = 'Close',title='Close price Analysis with Time Period ')

# update_xaxes() is used for updating the x-axis properties
# rangeslider_visible = True -> will put slider to arrange time duration
figure.update_xaxes(rangeslider_visible=True)

# update_xaxes(rangeselector=dict()) is used for create a range selector on the x-axis 
# The step parameter can be set to ( "month" | "year" | "day" | "hour" | "minute" | "second" | "all" )
# The stepmode parameter can be set to "backward" or "todate" to specify the direction of the range selector.
figure.update_xaxes(
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=3, label="3m", step="month", stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=1, label="YTD", step="year", stepmode="todate"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(step="all")
        ])
    )
)
figure.show() # show graph
```

```python
# candlestick chart for open, close, high, low trends in Microsoft prices(with filters for time period of 1m, 6m, YTD, 1y, all)

# Plotly.graph_objects() is used for creating interactive candlestick graphs
import plotly.graph_objects as go
# figure = px.candlestick(df, x=df.index, y='Close',title='Close price Analysis with Time Period Selectors')
figure = go.Figure(data=[go.Candlestick(x=MSFT_stock_data.index,open=MSFT_stock_data['Open'],high=MSFT_stock_data['High'],low=MSFT_stock_data['Low'],close=MSFT_stock_data['Close'])])

# update_xaxes() is used for updating the x-axis properties
# rangeslider_visible = True -> will put slider to arrange time duration
figure.update_xaxes(rangeslider_visible=True)

# update_xaxes(rangeselector=dict()) is used for create a range selector on the x-axis 
# The step parameter can be set to ( "month" | "year" | "day" | "hour" | "minute" | "second" | "all" )
# The stepmode parameter can be set to "backward" or "todate" to specify the direction of the range selector.
figure.update_xaxes(
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=3, label="3m", step="month", stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=1, label="YTD", step="year", stepmode="todate"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(step="all")
        ])
    )
)

# add labels on x, y axis and title
figure.update_layout(
    title = 'Close price Analysis with Time Period ',
    yaxis_title = 'Close Price',
    xaxis_title = 'Date'
)
figure.show() # show graph
```

**TIme Series Analysis:**

Time series analysis helps us to analyze the pattern of trends that happened in the past for particular period.

We can analyze data of particular day, week, month, year, etc. Which help us to get the daily return, monthly return.

It helps us to find the correlation between different set of data like Open price will increase than there is also possibility to increase high price, low price and close price.

We can also create future trends using time series.
